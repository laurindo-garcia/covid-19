# Generating a list of three digit numbers

```python
def mylist_a(r1, r2):
    if (r1 == r2):
        return r1
    else:
        res = []
        while (r1 < r2+1):
            res.append(r1)
            r1 += 1
        return res
        
r1, r2 = 100, 999
print(mylist_a(r1, r2))
```

```python
import numpy as np
prod = np.array(mylist_a(r1, r2))
```

```python
print(prod)
```

# Generating a list of number based on the product of three digit numbers

```python
prod2 = prod * prod
print(prod2)
```

# Covert list from integers to string

```python
six_digit = str(prod2)
```

```python
print(six_digit)
```
