# Generating a list of three digit numbers

```python
def mylist_a(r1, r2):
    if (r1 == r2):
        return r1
    else:
        res = []
        while (r1 < r2+1):
            res.append(r1)
            r1 += 1
        return res
        
r1, r2 = 100, 999
print(mylist_a(r1, r2))
```

```python
import numpy as np
prod = np.array(mylist_a(r1, r2))
```

```python
print(prod)
```

# Generating a list of number based on the product of three digit numbers

```python
prod2 = prod * prod
print(prod2)
```

# Covert list from integers to string

```python
six_digit = str(prod2)
```

```python
print(six_digit)
```

# Rethinking the approach (v2)

```python
# evaluate the product of two 3-digit nuumbers
# initialise variable for palindromes
# use a loop to evaluate if int(palindrome) is largest in series

palindrome = []
for i in range(100,999):
    for j in range(999, 100, -1):
        multi = i * j
        # find palindromes from product
        if str(multi) == str(multi)[::-1]:
            palindrome.append(multi)

answer = str(max(palindrome))
print("the largest palindrome max from the product of two 3-digit numbers is " + answer)
```
